```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def print_graphic(path):
    def plot_data(x, y, z, title, linestyle1='-', linestyle2=':'):
        plt.figure(figsize=(15, 9))  # Задайте нужные вам значения для width и height
        plt.plot(x, y, marker='o', linestyle=linestyle1, label='Проходы')
        plt.plot(x, z, marker='o', linestyle=linestyle2, label='Копирования')
        plt.title(title)
        plt.xlabel('X-axis Label')
        plt.ylabel('Y-axis Label')
        plt.legend()
        plt.show()

    x1, y1, z1 = np.loadtxt(f"rand_data_{path}_sort.txt", unpack=True)
    plot_data(x1, y1, z1, f'{path} rand_data', linestyle1='-', linestyle2=':')

    x2, y2, z2 = np.loadtxt(f"sorted_{path}.txt", unpack=True)
    plot_data(x2, y2, z2, f'{path} sorted', linestyle1=':', linestyle2='-')

    x3, y3, z3 = np.loadtxt(f"unsorted_{path}.txt", unpack=True)
    plot_data(x3, y3, z3, f'{path} unsorted', linestyle1=':', linestyle2='-')
```

```python
print_graphic('bubble')
```

```python
print_graphic('shaker')
```

```python
print_graphic('comb')
```
